<h3 style="color: plum;"><b>Python is an Object-Oriented Language</b></h3>


In Python, everything is an object.

That means these are all objects:

    integers -> 10
    strings -> "hello"
    lists -> [1,2,3]
    functions -> print, len
    classes -> str, list
    even modules -> math, os
---
So, in Python:

>Data + Behavior are represented using objects.

<h3><b>What is an Object in Python?</b></h3>

Definition (theoretical)

    An object is a runtime entity that has:
    1) Identity -> unique memory reference (id())
    2) Type -> what kind of object it is (type())
    3) Value / State -> actual data stored

Example:

```python
x = 50
print(id(x)) # identity  
print(type(x)) # type  
print(x) # value
```

Internally:

    1) 10 is an object of class int
    2) int itself is also an object (a class object)

 <h3 style="color: plum;"><b>Accessing and calling objects in Python</b></h3>

 dot operator(.) is an **attribute access**

when you write:

```python
"hello".upper()
```

python interprets:
1) "hello" -> object
2) upper -> attribute (method) stored inside the class str

so internally it becomes:

str.upper("hello")

Meaning:

>Methods are functions stored inside a class, and python passes the objcy automatically.

<h3 style="color: plum;"><b>Built-in functions vs Methods (Internal Working)</b></h3>

Example:

```python
len("python")
```

internal meaning:

Python searches `len` inside the built-in namespace.

built-in lives in:

`_builtins_` & `builtins`module

so `len("python")` is basically, `builtins.len(""python")`

built-in functions are not tied to one object.

---

<h3 style="color: plum;"><b>Method belongs to an object's class</b></h3>

example:




```python
"python".upper()
```

Internal meaning:

python does:

1) look at object type -> `str`

2) find `upper` inside `str`

3) connect it with the object "python"

4) then call it 

Equivalent internal call:

str.upper("python")

so, method = function that becomes bound to an object.

---

<h3 style="color: plum;"><b>Why methods feel different?</b></h3>

because in methods:

Python automatically passes the object as tge first argument

that object is called `self` (conventionally)

example:



```python
class demo:
    def show(self):
        print("Hello World")
d = demo()
d.show()
```

internally:
demo.show(d)

so, `d.show()` is just a shortcut for `demo.show(d)`
