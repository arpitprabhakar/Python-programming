# What is interactive programming in Python? (5 Marks)

Interactive programming in Python means running code directly in the Python interpreter (REPL mode) from the terminal.

You type python or python3, get the >>> prompt, write one line of code, press Enter, and see the result instantly. No file saving, no python script.py command needed.


# What is script programming in python? - (5 Marks)

Script programming in Python involves writing code in a .py file and executing it entirely using the terminal command python filename.py (or python3 filename.py). The interpreter reads and runs the whole script at once, unlike interactive mode where code runs line-by-line instantly. This mode is used for developing complete programs, as it allows saving, editing, and reusing code easily.

# Dynamic Types in Python

Python is a dynamically typed language, which means variables are created and assigned at runtime. You do not need to explicitly specify a data type when declaring a variable. Instead, Python automatically determines the datatype based on the value assigned to the variable, and this datatype can even change during program execution.


```python
a=30
```

```python
a
```

```python
a=20
print(type(a))
a=20/100
print(type(a))
a="Text"
print(type(a))
```

# What is a tuple?

A tuple is an ordered, immutable collection of elements. once a tuple is created, it can't change, add, or remove its items. Tuples are defined by enclosing comma-separated values in parentheses ()

# Varriable - Native Python?

a variable is simply a named location in memory used to store data. The value stored in a variable can change while the program is running, which makes variables flexible and useful for handling different kinds of data.

**Guidelines for naming and Using Variables:**

1) Variable names can include letters, numbers, and underscores. They must start with a letter or an underscore, not a number. For example, message_1 is valid, but 1_message is not.

2) Spaces are not allowed in variable names. Use underscores to separate words, like greeting_message instead of greeting message.

3) Do not use Python keywords or built-in function names as variables. Words like print are reserved for specific purposes in Python.

4) Choose names that are short but descriptive. For instance, name is clearer than n, and student_name is better than s_n.

5) Be cautious when using lowercase l and uppercase O, as they can easily be mistaken for the numbers 1 and 0
