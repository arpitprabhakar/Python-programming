<h4><b>What is interactive programming in Python? (5 Marks)</b></h4>

Interactive programming in Python means running code directly in the Python interpreter (REPL mode) from the terminal.

You type python or python3, get the >>> prompt, write one line of code, press Enter, and see the result instantly. No file saving, no python script.py command needed.


<h4><b>What is script programming in python? - (5 Marks)</b></h4>

Script programming in Python involves writing code in a .py file and executing it entirely using the terminal command python filename.py (or python3 filename.py). The interpreter reads and runs the whole script at once, unlike interactive mode where code runs line-by-line instantly. This mode is used for developing complete programs, as it allows saving, editing, and reusing code easily.

# Dynamic Types in Python

Python is a dynamically typed language, which means variables are created and assigned at runtime. You do not need to explicitly specify a data type when declaring a variable. Instead, Python automatically determines the datatype based on the value assigned to the variable, and this datatype can even change during program execution.


```python
a=30
```

**Guidelines for naming and Using Variables:**

1) Variable names can include letters, numbers, and underscores. They must start with a letter or an underscore, not a number. For example, message_1 is valid, but 1_message is not.

2) Spaces are not allowed in variable names. Use underscores to separate words, like greeting_message instead of greeting message.

3) Do not use Python keywords or built-in function names as variables. Words like print are reserved for specific purposes in Python.

4) Choose names that are short but descriptive. For instance, name is clearer than n, and student_name is better than s_n.

5) Be cautious when using lowercase l and uppercase O, as they can easily be mistaken for the numbers 1 and 0

**1) Case Conversion Methods**

<br>

>lower()

<h4><b>Definition and Usage</b></h4>

Return a copy of the string with all the cased characters converted to lowercase. 

For example:

```python
"PYTHON".lower()
```

<hr>
<br>

>upper()

<h4><b>Definition and Usage</b></h4>

Returns a copy of the string with all characters converted to uppercase.

Note: s.upper().isupper() might be False if s contains uncased characters or if the Unicode category of the resulting character(s) is not “Lu” (Letter, uppercase), but e.g. “Lt” (Letter, titlecase).

For example:

```python
"python".upper()
```

<hr>
<br>

>title()

<h4><b>Definition and Usage</b></h4>

Return a title cased version of the string where words start with an uppercase character and the remaining characters are lowercase.

For example:

```python
"python programming".title()
```

<hr>
<br>

>capitalize()

<h4><b>Definition and Usage</b></h4>

The capitalize() Return a copy of the string with its first character capitalized and the rest lowercased.

*NOTE: Changed in version 3.8: The first character is now put into titlecase rather than uppercase. This means that characters like digraphs will only have their first letter capitalized, instead of the full character.*

For example:

```python
"python programming".capitalize()
```

<hr>
<br>

>swapcase()

<h4><b>Definition and Usage</b></h4>

The swapcase() method returns a string where all the upper case letters are lower case and vice versa

For example:

```python
"PyThOn".swapcase()
```

<hr>
<br>

>casefold()

<h4><b>Definition and Usage</b></h4>

Returns a casefolded version of the string, which is mainly used for caseless string matching.

Casefolding is similar to lowercasing but more aggressive because it is intended to remove all case distinctions in a string. For example, the German lowercase letter 'ß' is equivalent to "ss". Since it is already lowercase, lower() would do nothing to 'ß'; casefold() converts it to "ss".

For example:

```python
"GERMAN ß".casefold()
```

**2) Remove Spaces/ cleanup**

<br>

>strip(chars=None)

<h4><b>Definition and Usage</b></h4>

Return a copy of the string with the leading and trailing characters removed. The chars argument is a string specifying the set of characters to be removed. If omitted or None, the chars argument defaults to removing whitespace. The chars argument is not a prefix or suffix; rather, all combinations of its values are stripped 

For example:

```python
"   hello   ".strip()   #   'hello'
```

<hr>
<br>

>lstrip(chars=None)

<h4><b>Definition and Usage</b></h4>

Return a copy of the string with leading characters removed. The chars argument is a string specifying the set of characters to be removed. If omitted or None, the chars argument defaults to removing whitespace. The chars argument is not a prefix; rather, all combinations of its values are stripped:

For example:

```python
"   hello   ".lstrip()     # 'hello'
```

<hr>
<br>

>rstrip(chars=None)

<h4><b>Definition and Usage</b></h4>

Return a copy of the string with trailing characters removed. The chars argument is a string specifying the set of characters to be removed. If omitted or None, the chars argument defaults to removing whitespace. The chars argument is not a suffix; rather, all combinations of its values are stripped:

For example:

```python
"   hello   ".rstrip()     # 'hello'
```
